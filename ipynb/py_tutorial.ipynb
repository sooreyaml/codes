```python
squares = ["red", "yellow", "blue",'white','white']

for i in range(0, len(squares)):
    squares[0] = "white"
    print(squares)
```

```python
for square in squares:
    print(square)
```

## reading files

```python
tweets = open("tweets.json","r")
# w : to write
# r : to read
# a : to append
```

```python
# using a with statement automatically closes the file.
with open("basics.txt","r") as file1:
    file_stuff = file1.read(1)
    print(file_stuff) 
print(file1.closed)
print(file_stuff)
```

```python
# using a with statement automatically closes the file.
with open("write.txt","w") as file1:
    file1.write("This is a file that is ambiguous.")
print(file1.closed)
```

```python
lines = ["This is line One.\n", "This is line Two.\n", "This is line Three."]
with open("write.txt","w") as file1:
    for line in lines:
        file1.write(line)
```

```python
with open("write.txt","a") as file1:
    file1.write("\nThis is line Four.")
```

```python
# to copy from one file to another.
with open("basics.txt", "r") as readFile:
    with open("write.txt", "a") as writtenFile:
        for line in readFile:
            writtenFile.write(line)
```
